## <i>Рисунки и графики</i>

```python
import numpy as np
```

```python
array_x = np.linspace(0, 5, 20)
array_y = (5)*np.random.random((20,))
array_y_2 = np.random.uniform(0,5,(20))
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 12,9
```

```python
fig, ax = plt.subplots()
ax.plot(array_x,array_y, marker = 'o', c = 'goldenrod', linewidth = 4, label='Рандом обычный от 0 до 5')
ax.plot(array_x,array_y_2, c = 'navy', linestyle = '-.', linewidth = 2.5, label='Рандом uniform от 0 до 5')
ax.set_title('Графики рандомных чисел от 0 до 5', fontsize= 20)
ax.set_xlabel('Ось X', fontsize= 20)
ax.set_ylabel('Ось Y', fontsize= 20)
ax.legend(loc='upper right')
```

```python
fig, ax = plt.subplots()
N = 100
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = (30 * np.random.rand(N))**2

ax.scatter(x, y, s=area, c=colors, alpha=0.5, marker='o', cmap='cividis')
```

```python
fig, ax = plt.subplots()

np.random.seed(19680801)

mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

ax.hist(x, 50, density=True, facecolor='g', alpha=0.75)
```
